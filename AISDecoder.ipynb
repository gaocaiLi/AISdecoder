Reading and decoding a file 

```python
#initializing parsing signs
parse_sign = '!'
time_stamp_split = ' '
line_split = ','

#initializing variables needed for counting lines
count_lines = 0
not_time_stamp = 0
time_stamp = 0
no_of_exeptions = 0

#creating a .txt file to write all decoded messages in 
output = open("decoded.txt", "w+")

#opening one file                                      NOTE: CHECK HOW TO OPEN THE WHOLE FOLDER 
with gzip.open('./ais-10-10/23-59.txt.gz', 'rb') as f:
    for line in f:
        count_lines += 1
        try:
            #For all lines not starting with a time stamp
            if line.startswith(parse_sign):
                not_time_stamp += 1
                decoded = str(ais.decode(line.split(line_split)[5], int(line.split(line_split)[6][:1])))
                output.write(decoded + '\n\n')
                #print decoded
                
            #For the lines starting with a time stamp
            else:
                time_stamp += 1
                list1 = line.split(time_stamp_split)
                timestamp = datetime.datetime.fromtimestamp(float(list1[0]))
                output.write(str(timestamp) + '\n\n')
                #print('\n\n' + timestamp + '\n\n')
                try:
                    decoded = str(ais.decode(line.split("line_split")[5], int(line.split("line_split")[6][:1])))  
                    output.write(decoded + '\n\n')
                    #print(decoded + '\n\n')
                except Exception:
                    no_of_exeptions +=1  
                    pass
                
                
        #catches all the uncommon messages  
        except Exception:
            no_of_exeptions +=1
            print line
            pass
            
#output.colse()
```
